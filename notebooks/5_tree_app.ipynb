```python
import nbappinator as nb
```

```python
def get_tree_value(app):
    val = app["tree1"]
    app.done(f"Tree selected value = {val}")
```

```python
# Create app with tree widget
myapp = nb.App(tabs=["Tree Widget"])

mainpage = myapp.tab(0)

mainpage.label("The following is a selectable tree view")

# Each node must be added in order: can't add org1/something without a preceding org1, for instance.
mainpage.tree("tree1", paths=["org1", "org1/something", "org2", "org2/somethingelse"], delimiter="/")
mainpage.button("get_val", on_click=get_tree_value, label="Get Selected Value from Tree", status=True)

selectedval = myapp["tree1"]

myapp.display()
```

```python
# Select some nodes programmatically

t = myapp._widgets["tree1"]

# With D3Tree, set selected paths directly
t.selected = ["org1", "org2"]

selectedval = myapp["tree1"]
# Use set comparison since order may vary
assert set(selectedval) == {"org1", "org2"}
```

```python
# Second tree example
myapp2 = nb.App(tabs=["Tree Widget"])

mainpage2 = myapp2.tab(0)

mainpage2.label("The following is a selectable tree view")

# Each node must be added in order: can't add org1/something without a preceding org1, for instance.
mainpage2.tree("tree1", paths=["org1", "org1/something", "org2", "org2/somethingelse", "boo/boo"], delimiter="/")
mainpage2.button("get_val", on_click=get_tree_value, label="Get Selected Value from Tree", status=True)

selectedval = myapp2["tree1"]

myapp2.display()
```
