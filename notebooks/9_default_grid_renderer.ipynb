# Default Grid Renderer for DataFrames

Use `register_grid_renderer()` to make AG Grid the default renderer for all pandas DataFrames in your notebook.

```python
import pandas as pd

# Sample data
df = pd.DataFrame(
    {
        "name": ["Apple", "Microsoft", "Google", "Amazon", "Meta"],
        "price": [178.50, 378.91, 141.80, 178.25, 505.95],
        "volume": [52_340_000, 18_920_000, 21_450_000, 45_670_000, 12_340_000],
        "market_cap": [2.8e12, 2.9e12, 1.8e12, 1.9e12, 1.3e12],
    }
)
```

## Default pandas rendering

Before registering, DataFrames render as HTML tables:

```python
df
```

## Register AG Grid as default renderer

```python
from nbappinator import register_grid_renderer

register_grid_renderer(height=250)
```

## Now DataFrames render as AG Grid

```python
df
```

```python
# Any DataFrame automatically uses AG Grid
pd.DataFrame({"a": range(10), "b": range(10, 20)})
```

## Restore default rendering

Use `unregister_grid_renderer()` to go back to HTML tables:

```python
from nbappinator import unregister_grid_renderer

unregister_grid_renderer()
```

```python
# Back to HTML table
df
```
